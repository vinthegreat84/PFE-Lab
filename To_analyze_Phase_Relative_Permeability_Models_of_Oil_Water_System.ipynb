To analyze Phase Relative Permeability Models of Oil-Water System

1. **Corey's 2 Phase Model - Valid for Drainage in a Consolidated Rock.**

$Swn = \frac {Sw - Swirr}{1 - Swirr}  $

$Krw = Swn^4$

$Kro = (1 - Swn)^2 * (1 - Swn^2)$

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
Sw = np.linspace(0,1,100)

Swirr = 0.20

Swn = (Sw - Swirr)/(1-Swirr)

Krw = Swn**4

Kro = ((1 - Swn)**2)*(1 - Swn**2)

plt.style.use('default')
plt.plot(Sw,Krw,Sw,Kro,label=['Krel-water(Wetting P.)','Krel-Oil(Non-Wetting P.)'],linewidth=3)
plt.title("Rel-Perm by Corey's Model")
plt.xlabel('Sw')
plt.ylabel('Krel(o,w)')
plt.grid()
plt.legend(loc='best')
```
