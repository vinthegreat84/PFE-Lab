```python
# function for the measurement of mudcake thickness
def mudcake_dia (d_h):
  d_bit = 8.5 # as read from the log; change the value as and when applicable
  h_mc  = (d_bit - d_h) / 2
  return h_mc
```

```python
print('5.1 Mean mud-cake thickness in interval 2980 m to 3035 m, hmc =', mudcake_dia(7.5), 'in') # caliper dia taken from the log in interval 2980 m to 3035 m
print('5.2 Mean mud-cake thickness in interval 3082 m to 3145 m, hmc =', mudcake_dia(8.0), 'in') # caliper dia taken from the log in interval 3082 m to 3145 m
```

```python
# function for the measurement of borehole volume
def borehole_vol (d_h):
  V_h  = d_h**2/2
  V_h  = V_h + V_h*0.012
  return V_h
```

```python
start_depth = 2900 # as read from the log; change the value as and when applicable 
end_depth   = 3250 # as read from the log; change the value as and when applicable

print('6. The approximate borehole volume in the intervals', start_depth, 'to', end_depth, 'm,', 'Vh =', borehole_vol (7.5), 'litres/m', 'or', 
borehole_vol (7.5) * (end_depth - start_depth), 'litres')
```

```python
# function for the measurement of required cement volume
def cement_vol (d_h, d_casing):
  V_cement  = 0.5 * (d_h**2 - d_casing**2)
  V_cement  = V_cement + V_cement*0.01
  return V_cement
```

```python
print('7. The cement volume required to complete the borehole between', start_depth, 'to', end_depth, 'm,', 'Vcement =', cement_vol (7.5, 7.0), 'litres/m', 'or', 
cement_vol (7.5, 7.0) * (end_depth - start_depth), 'litres')
```
