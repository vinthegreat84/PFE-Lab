> # **Problem Statement**

The following exercise illustrates the adjustment of fluid resistivities for temperature as described in *Asquith et. al., 2004*.


**Asquith, G.B., Krygowski, D. and Gibson, C.R.**, 2004. Basic well log analysis (Vol. 16). Tulsa: American Association of Petroleum Geologists, pp. 19.

The formation temperature is calculated (Asquith, 1980) by using the linear regression equation:

$ y = mx + c $

where:

x = depth

y = temperature

m = slope (In this example it is the geothermal gradient.)

c = a constant (In this example it is the mean annual surface temperature.)

**Problem:**

Surface temperature = 80°F

Bottom hole temperature (BHT) = 200°F

Total depth (TD) = 10,000 ft

Formation depth = 7000 feet

> ## **Solution**

**Method - 1:**

```python
# Surface temperature = 80°F
ST = 80

# Bottom hole temperature (BHT) = 200°F
BHT = 200

# Total depth (TD) = 10,000 ft
TD = 10000

# Formation depth = 7000 feet
FD = 7000

# slope (the geothermal gradient)
m = (BHT - ST) / TD

# The formation temperature, FT
FT = m * FD + ST

print('The formation temperature, FT = ', round(FT, 0), '°F')
```
